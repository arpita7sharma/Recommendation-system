```python
book_pivot = book.pivot_table(index ='userid', columns ='title', values="rating").reset_index(drop=True)
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

```python
book_pivot.index = book.userid.unique()
```

```python
book_pivot
```

```python
book_pivot.isna().sum()
```

```python
#Calculating Cosine Similarity between Users
user_sim = 1 - pairwise_distances( book_pivot.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
#Set the index and column names to user ids 
user_sim_df.index = book.userid.unique()
user_sim_df.columns = book.userid.unique()
```

```python
user_sim_df.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```

```python
#Most Similar Users
print(user_sim_df.idxmax(axis=1)[1348])
print(user_sim_df.max(axis=1).sort_values(ascending=False).head(50))
```
